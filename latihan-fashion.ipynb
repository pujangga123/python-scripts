````python
import pandas as pd

df = pd.read_csv("fashion.csv", sep=";")
df
````

Menampilkan total profit di bulan januari, output:

```
Profit Januari = ....
Kesimpulan = SESUAI TARGET
```


````python
januaridata = df[df.Month==1]
total = januaridata.Profit.sum()
print("Profit Januari = ", total)

if total>2000:
  print("Kesimpulan = SESUAI TARGET")
else:
  print("Kesimpulan = TIDAK SESUAI TARGET")
````

Tampilkan profit untuk setiap bulan ganjil

```
Profit bulan 1 = ...
Profit bulan 3 = ...
Profit bulan 5 = ...
...
```

````python
for m in range(1,13,2):
  databulan = df[df.Month==m]
  totalprofit = databulan.Profit.sum()
  print("Profit bulan",m,"=",df[df.Month==m].sum())

  # alternatif
  # print("Profit bulan",m,"=",df[df.Month==m].Profit.sum())
````

````python
m = 1
while  m < 13:
  databulan = df[df.Month==m]
  totalprofit = databulan.Profit.sum()
  print("Profit bulan",m,"=",totalprofit)
  m = m + 2
````

Tampilan total profit dari script sebelumnya, tapi dengan menampilkan bulan dengan nama bulan
```
Profit bulan Januari = ...
Profit bulan Maret = ...
Profit bulan Mei = ...
...
```

````python
namabulan = ["Januari","Februari", "Maret", "April","Mei","Juni",
             "Juli","Agustus","September","Oktober","November","Desember"]
jumlah = 0
for m in range(1,13,2):
  databulan = df[df.Month==m]
  totalprofit = databulan.Profit.sum()
  print("Profit bulan", namabulan[m-1] ,"=",round(totalprofit,1))
  jumlah = jumlah + totalprofit

print("Total =", round(jumlah,1))
````
